```python
import pandas as pd
from datasets import Dataset

# Load your Kaggle CSV
df = pd.read_csv("/content/student_productivity_distraction_dataset_20000.csv")

def format_row(row):
    instruction = f"Analyze the productivity of a student with {row['study_hours_per_day']}h study, {row['sleep_hours']}h sleep, and {row['phone_usage_hours']}h screen time."
    response = f"This student has a productivity score of {row['productivity_score']}/10. Analysis: High screen time is likely impacting focus. Suggestion: Reduce digital distraction by 20%."
    return {"instruction": instruction, "output": response}

# Convert to HuggingFace Dataset format
dataset = Dataset.from_pandas(df.apply(format_row, axis=1, result_type='expand'))
```

```python
from unsloth import FastLanguageModel
import torch

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/llama-3-8b-bnb-4bit", # Highly efficient 4-bit version
    max_seq_length = 2048,
    load_in_4bit = True,
)

# Explicitly set the chat template for Llama-3
# This is the recommended chat template for Llama-3 models.
tokenizer.chat_template = """{% for message in messages %}{% if message['role'] == 'user' %}{{ '<|start_header_id|>user<|end_header_id|>
' + message['content'] | trim + '<|eot_id|>' }}{% elif message['role'] == 'system' %}{{ '<|start_header_id|>system<|end_header_id|>
' + message['content'] | trim + '<|eot_id|>' }}{% elif message['role'] == 'assistant' %}{{ '<|start_header_id|>assistant<|end_header_id|>
' + message['content'] | trim + '<|eot_id|>' }}{% endif %}{% if loop.last and add_generation_prompt %}{{ '<|start_header_id|>assistant<|end_header_id|>
' }}{% endif %}{% endfor %}"""

# Add LoRA Adapters (The actual 'tuning' part)
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Rank: higher = more smarts but more memory
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj"],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
)
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments

def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    outputs      = examples["output"]
    texts = []
    for instruction, output in zip(instructions, outputs):
        # Use the tokenizer's chat template to format the instruction and output
        text = tokenizer.apply_chat_template(
            [{"role": "user", "content": instruction}, {"role": "assistant", "content": output}],
            tokenize=False,
            add_special_tokens=False
        ) + tokenizer.eos_token # Add EOS token at the end of each example
        texts.append(text)
    return texts # Changed to return a list of strings directly

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    # Remove dataset_text_field as we're now using a formatting_func
    formatting_func = formatting_prompts_func,
    max_seq_length = 2048,
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 2500, # Increased for better accuracy
        learning_rate = 2e-4,
        fp16 = not torch.cuda.is_bf16_supported(),
        logging_steps = 10,
        output_dir = "outputs",
    ),
)

trainer.train()
```

```python
# Enable fast inference
FastLanguageModel.for_inference(model)

# Use the exact same format as your training loop
prompt = """Analyze student productivity.
Rules: The score MUST be between 0 and 10.
Data: 10h study, 8h sleep, 4h screen time.
Analysis:"""
formatted_prompt = f"### Instruction:\n{prompt}\n\n### Response:\n"

inputs = tokenizer([formatted_prompt], return_tensors="pt").to("cuda")

outputs = model.generate(
    **inputs,
    max_new_tokens=64,
    temperature=0.1,  # Lower temperature = more deterministic, less random
    top_p=0.9,       # Focuses on the most likely tokens
    use_cache=True
)

# Decode only the NEW part (the response)
decoded = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
final_response = decoded.split("### Response:\n")[-1]
print(final_response)
```
